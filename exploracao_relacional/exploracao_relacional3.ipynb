```python
from nycflights13 import planes
```

```python
planes_df = pd.DataFrame(planes)
```

```python
planes_df.head(10)
```

```python
len(planes_df['tailnum']) == len(planes_df['tailnum'].unique())
```

```python
flights_planes_df = pd.merge(flights_df[['tailnum', 'carrier', 'flight']], planes_df[['tailnum', 'manufacturer', 'model']], how='inner')
```

```python
flights_planes_df.head(10)
```

```python
len(flights_planes_df.tailnum.unique())
```

```python
operated_planes_manufacturer = flights_planes_df[['manufacturer', 'flight']].groupby(['manufacturer']).count().sort_values('flight', ascending=True)
```
