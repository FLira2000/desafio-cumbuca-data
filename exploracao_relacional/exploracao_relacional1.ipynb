```python
from nycflights13 import flights
```

```python
from matplotlib import pyplot as plt
```

```python
flights.head(10)
```

```python
import pandas as pd
```

```python
flights_df = pd.DataFrame(flights)
```

```python
flights_df['delayed'] = flights_df['dep_delay'] > 0
```

```python
daily_delays = flights_df.groupby(['year', 'month', 'day'])['delayed'].mean().reset_index()
```

```python
moving_avg = daily_delays.rolling(window=30).mean()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(moving_avg)
plt.title('Moving Average of Delayed Flights')
plt.xlabel('Days')
plt.ylabel('Proportion of Flights Delayed')
plt.show()
```

```python
rolling_std = daily_delays['delayed'].rolling(window=30).std(ddof=0)
```

```python
plt.plot(rolling_std)
plt.title('Rolling Standard Deviation of Delayed Flights')
plt.xlabel('Days')
plt.ylabel('Proportion of Flights Delayed')
plt.show()
```
