```python
from nycflights13 import flights
```

```python
import pandas as pd
```

```python
flights_df = pd.DataFrame(flights)
```

```python
flights_df['delayed'] = flights_df['dep_delay'] > 5
```

```python
flights_df
```

```python
delayed_flights = flights_df[['year', 'month', 'carrier', 'delayed']].query('delayed == True').groupby(['year', 'month', 'carrier']).count()['delayed']
```

```python
total_flights = flights_df[['year', 'month', 'carrier', 'delayed']].groupby(['year', 'month', 'carrier']).count()['delayed']
```

```python
total_flights.columns = ['total']
```

```python
total_merged_flights = pd.merge(total_flights, delayed_flights, on=['year', 'month', 'carrier'])
```

```python
total_merged_flights.columns = ['total', 'delayed']
```

```python
total_merged_flights['percentual'] = total_merged_flights.apply(lambda r: (r.delayed / r.total) * 100, axis=1)
```
